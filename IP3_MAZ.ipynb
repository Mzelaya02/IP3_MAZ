### 4.	Import car-sales.csv  into a DataFrame called car_sales and view.

```python
car_sales=pd.read_csv('car-sales.csv')
car_sales.head()
```

### a.	Try to plot the 'Price' column using the plot() function

```python
fig,ax3=plt.subplots()
ax3.plot(car_sales['Price'])
ax3.set_ylabel('Price',color='Blue',size='large')
```

### b.	Try to plot the ''Odometer (KM) column using the plot() function

```python
# Both examples worked for me
fig,ax4=plt.subplots()
ax4.plot(car_sales['Odometer (KM)'])
ax4.set_ylabel('Odometer (KM)',color='Blue',size='large')
```

### 5.	Remove all the symbols from the 'Price' column and convert  and update it to  64-bit int numbers using astype(int64) Hint. Look for string functions like replace/strip).

```python
car_sales['Price']=car_sales['Price'].str.replace('$','').str.replace(',','').astype('int64')
car_sales.head()
```

### 6. Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
car_sales['Total Sales']=np.cumsum(car_sales['Price'])
car_sales.head()
```

### 7. Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out date_range function and keep the periods=len(car_sales). 

```python
car_sales['Sale Date']=pd.date_range(dt.today(), periods=len(car_sales))
car_sales.head()
```

### 8.	Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels).

```python
fig,ax5= plt.subplots()
plt.title('Price vs. Total Sales')
ax5.set_xlabel('Price',color='Blue',size='large')
ax5.set_ylabel('Total Sales',color='Blue',size='large')
ax5.plot(car_sales['Price'],car_sales['Total Sales'])
```

### 9.	Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.

```python
fig,ax6=plt.subplots()
plt.title('Odometer vs. Price')
ax6.set_xlabel('Odometer (KM)',color='blue',size='large')
ax6.set_ylabel('Price',color='blue',size='large')
ax6.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
```

### 10.	Create a matplotlib bar graph of the 'Make' vs 'Odometer (KM)' columns

```python
fig,ax7=plt.subplots()
plt.title('Make vs. Odometer')
ax7.set_xlabel('Make',color='blue',size='large')
ax7.set_ylabel('Odometer (KM)',color='blue',size='large')
ax7.bar(car_sales['Make'],car_sales['Odometer (KM)'])
```

### 11.	Create a histogram of the 'Make' column

```python
fig,ax8=plt.subplots()
plt.title('Car Brands Made')
ax8.set_xlabel('Brands Made',color='blue',size='large')
ax8.hist(car_sales['Make'])
```

### 12.	Create a subplot of multiple plots (steps 8,9,10,11)

```python
fig=plt.figure(figsize=(20,12))

ax5=fig.add_subplot(221)
plt.title('Price vs. Total Sales')
ax5.set_xlabel('Price',color='Blue',size='large')
ax5.set_ylabel('Total Sales',color='Blue',size='large')
ax5.plot(car_sales['Price'],car_sales['Total Sales'])

ax6=fig.add_subplot(222)
plt.title('Odometer vs. Price')
ax6.set_xlabel('Odometer (KM)',color='blue',size='large')
ax6.set_ylabel('Price',color='blue',size='large')
ax6.scatter(car_sales['Odometer (KM)'],car_sales['Price'])

ax7=fig.add_subplot(223)
plt.title('Make vs. Odometer')
ax7.set_xlabel('Make',color='blue',size='large')
ax7.set_ylabel('Odometer (KM)',color='blue',size='large')
ax7.bar(car_sales['Make'],car_sales['Odometer (KM)'])

ax8=fig.add_subplot(224)
plt.title('Car Brands Made')
ax8.set_xlabel('Brands Made',color='blue',size='large')
ax8.hist(car_sales['Make'])

```

### 16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 

```python
df=pd.DataFrame(car_sales,columns=['Total Sales','Price','Odometer (KM)'])
sns.heatmap(df.corr(),annot=True,annot_kws={'size':30},linewidth=1)
plt.show()
```

##### The correlation between odometer and price is a relatively strong inverse correlation. While, the correlation between odometer and total sales is a fairly strong relationship. Lastly, the correlation between price and total sales is a relatively weak correlation.
